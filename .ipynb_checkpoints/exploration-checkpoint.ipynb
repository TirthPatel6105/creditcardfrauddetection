```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("C:/Users/tirth/Desktop/ML project latest/transactions.csv")
print(df.shape)
print(df['isFraud'].value_counts(normalize=True))

df['transactionDateTime'] = pd.to_datetime(df['transactionDateTime'])
df['accountOpenDate'] = pd.to_datetime(df['accountOpenDate'])
df['dateOfLastAddressChange'] = pd.to_datetime(df['dateOfLastAddressChange'])

# Features: time since account open / last address change
df['days_since_account_open'] = (df['transactionDateTime'] - df['accountOpenDate']).dt.days
df['days_since_address_change'] = (df['transactionDateTime'] - df['dateOfLastAddressChange']).dt.days

# Hour and weekday of transaction
df['trans_hour'] = df['transactionDateTime'].dt.hour
df['trans_dayofweek'] = df['transactionDateTime'].dt.dayofweek

print(df.groupby('transactionType')['isFraud'].mean())
df = df.drop(columns=['accountNumber', 'cardLast4Digits', 'merchantCountryCode', 'posConditionCode'])
df = pd.get_dummies(df, columns=['merchantName', 'transactionType', 'acqCountry', 'posEntryMode', 'cardPresent', 'expirationDateKeyInMatch'])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['transactionAmount'] = scaler.fit_transform(df[['transactionAmount']])

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

X = df.drop('isFraud', axis=1)
y = df['isFraud']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
model = RandomForestClassifier(class_weight='balanced', random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
